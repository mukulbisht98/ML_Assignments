```python
sns.distplot(data['casual'])
plt.title("casual ")
plt.show()
```

```python
sns.distplot(data1['casual'])
plt.title("casual ")
plt.show()
```

```python
sns.distplot(data['registered' ])
plt.title("registered")
plt.show()
```

```python
sns.distplot(data1['registered' ])
plt.title("registered")
plt.show()
```

```python
sns.distplot(data['weathersit' ])
plt.title("weathersit")
plt.show()
```

```python
plt.hist(data['workingday'],bins=10 )
plt.title('working day')
plt.xlabel('work')
plt.ylabel('day')
```

```python
plt.hist(data1['workingday'],bins=10 )
plt.title('working day')
plt.xlabel('work')
plt.ylabel('day')
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.corr(),annot=True)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(data1.corr(),annot=True)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(x='weekday', y='holiday', data=data, x_jitter=0.3, scatter_kws={'alpha':0.2})
sns.despine()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(x='weekday', y='holiday', data=data1, x_jitter=0.3, scatter_kws={'alpha':0.2})
sns.despine()
```
